NAVIGATING DATAFRAMES

Use ***loc*** or ***iloc*** to select one or more rows or values:

```
#load library
import pandas as pd

#Create URL
url ='https://raw.githubusercontent.com/chrisalbon/simulated_datasets/master/titanic.csv'

#Load data
df =pd.read_csv(url)

# Select first row
df.iloc[0]
```

```
#to select a slice of rows we want,lets say row 2 to row 7
#row2 =1
#row7 = 6

#always remember that Python starts counting from zero not one

df.iloc[1:6]
```

```
#to select a number of rows,let say 4 rows
df.iloc[:4]
```

DataFrames do not need to be numerically indexed. We can set the index of a DataFrame to any value where the value is unique to each row.
For example,we can set the index to be  passenger names and then select rows using a name

```

#Select by name
#Select index
df= df.set_index(df['Name'])


#Show row
df.loc['Allen, Miss Elisabeth Walton']
```

```
#Select the survivors
#Select index
df= df.set_index(df['Survived'])


#Show row
df.loc[1]
```

```
#Select by Sex
#Select index
df= df.set_index(df['Sex'])

#Show row
df.loc['female']
```

All rows in a pandas DataFrames has a unique index value. By default,this index is an integer indicating its row position in the DataFrame, however it does not have to be. DataFrame indexes can be set to be ***unique alphanumeric strings or customer numbers***. To select individual rows and slices of rows,pandas provides two methods: 



*   **loc** is useful when the index of the DataFrame is a Label (e.g a string)
*   **iloc** works by looking for the position in the DataFrame. For example, iloc[0] will return the first row regardless of whether the index is an integer or a label.

it is useful to be comfortable with both loc and iloc since it will come up with a lot during data

**Finding the Minimum, Maximum, Sum, Average, and Count**

```
# Load library
import pandas as pd

# Create URL
url = 'https://tinyurl.com/titanic-csv'

# Load data
justina = pd.read_csv(url)
# Calculate statistics
print('Maximum:', justina['Age'].max())
print('Minimum:', justina['Age'].min())
print('Mean:', justina['Age'].mean())
print('Sum:', justina['Age'].sum())
print('Count:', justina['Age'].count())
```

the syntax for doing the above is 
print('a:' , b['c'].d( ) )

where a = a text referring to something
            b = name of the file,in this case it is justina
            c = the particular column you want to do some statistics on(in this case we have **age**)
            d = the statistical method you wish to implement(it could be mean,max etc)
            
            to illustrate this better, we will do this example
            supposing **a** = calculated mode
                                  ** b** = justina
                                  ** c** = Age
                                   **d** = mode
                                   
         the code will be like the one below

```
#print('a:' , b['c'].d( ) )  use this as the syntax

print('calculated mode:' , justina['Age'].mode( ) )
```
